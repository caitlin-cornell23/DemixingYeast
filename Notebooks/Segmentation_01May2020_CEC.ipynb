# Segmentation of Vacuoles from Background #

### This is a draft notebook for a segmentation function ###

Import all of the necessary packages

```python
import numpy as np
import matplotlib.pyplot as plt

## Image manipulation
import skimage.segmentation as seg
from skimage.feature import canny
from skimage.filters import sobel, rank
from skimage import exposure
from skimage import img_as_float
from skimage.morphology import square
```

Import the image

```python
root_dir = "C://Users/caitl//Documents//GitHub//DemixingYeast//Data//"
im = img_as_float(plt.imread(root_dir + "Test.tif"))
```

Detect the edges to differentiate from background

```python
## Detect the edges with a sobel filter
im_filt = canny(im, sigma=3, low_threshold=0.01*im.max(), high_threshold=0.1*im.max())

figs, axes = plt.subplots(ncols=2, figsize=(10,5))

for ax in axes:
    ax.axis('off')
    
axes[0].imshow(im, cmap='gray')
axes[1].imshow(im_filt, cmap='gray')
```
